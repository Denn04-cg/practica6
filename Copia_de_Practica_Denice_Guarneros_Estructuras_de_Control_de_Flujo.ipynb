## **SELECCIÓN**

Una **estructura de selección** es una **declaración condicional**, el cual permite que diferentes secuencias de código se ejecuten dependiendo de si una **condición booleana** especificada se evalúa como **verdadera o falsa.**

La sintaxis básica de una condicional es la siguiente:

```
if condicion1:
    # código a ejecutar si la condicion1 es verdadera
elif condicion2: (opcional)
    # código a ejecutar si la condicion1 es falsa pero la condicion2 es verdadera
else:
    # código a ejecutar si tanto la condicion1 como la condicion2 son falsas
```

### **EJEMPLO 1:**

````python
# Definir variable
calificacion = int(input('Inserta una calificación: '))
````

````python
calificacion
````

````python
# Definir condicionales consecutivos (Independientes)
if calificacion == 100:
  print('Calificación perfecta')

if calificacion >= 80 and calificacion < 100:
  print('Calificación deseable')

if calificacion >= 60 and calificacion < 80:
  print('Calificación aprobatoria')
else:
  print('Calificación no aprobatoria')
````

````python
# Definir condicional compuesta (Elif)
if calificacion == 100:
  print('Calificación perfecta')
elif calificacion >= 80 and calificacion < 100:
  print('Calificación deseable')
elif calificacion >= 60 and calificacion < 80:
  print('Calificación aprobatoria')
else:
  print('Calificación no aprobatoria')
````

````python
# Definir condicionales anidadas
if calificacion == 100:
  print('Calificación perfecta')
  if calificacion >= 80 and calificacion < 100:
    print('Calificación deseable')
    if calificacion >= 60 and calificacion < 80:
      print('Calificación aprobatoria')
    else:
      print('Calificación no aprobatoria')
````

````python
"""
== -> Igual a
!= -> Es diferente de
>  -> Mayor a
<  -> Menor a
>= -> Mayor igual a
<= -> Menor igual a
"""
````

## **ITERACIÓN**

En Python, una **estructura de iteración** es un **ciclo** que generalmente se refiere a dos tipos de **estructuras de control de flujo**: los bucles **for** y **while**.


*   **Bucle For**: El bucle for se usa para **recorrer sobre una secuencia** (que puede ser una lista, una tupla, un diccionario, un conjunto o una cadena) **o cualquier otro objeto iterable**. Cada elemento de la secuencia se procesa en el cuerpo del bucle, uno tras otro.

*   **Bucle While**: El bucle while se usa para **ejecutar repetidamente un bloque de código** siempre que una condición especificada sea verdadera.

### **CICLO FOR**

También conocido como **bucle for**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de código un número determinado de veces.

El ciclo for tiene la siguiente estructura:

```
for variable in secuencia:
    # cuerpo del ciclo

```

### **CICLO WHILE**

También conocido como **bucle while**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de código mientras se cumpla una condición específica.

El ciclo for tiene la siguiente estructura:

```
while condicion:
    # código a ejecutar mientras se cumpla la condición
```

Recorrer las opciones de un menú:

````python
# Definir variables
menu = '''
********** MENU ********** \n
1) Mi primer "Hola, mundo!"
2) Factorial de 5
3) Salir del menú
'''

print(menu)
````

````python
# Realizar ciclo while
flag_menu = True

while True:
  print(menu)
  opcion = int(input('Ingresa una opción: '))

  if opcion == 1:
    print('Hola, mundo!')

  elif opcion == 2:
    factorial = 1

    for numero in range(1, 6):
      factorial *= numero
      print(factorial)

  elif opcion == 3:
    print('Saliendo del menú...')
    break
  else:
    print('Opción no encontrada')
````

#### **EJEMPLO 3:**

Calcular la suma de los valores de un diccionario:

````python
# define tus variables
vocales = {'a': 1, 'e': 2, 'i': 3, 'o': 4, 'u': 5}
vocal = input('Inserta una vocal: ')

claves = list(vocales.keys())
limite = claves.index(vocal)

print(claves)
print(limite)

lista = []
indice = 0
suma = 0
````

````python
# realiza el ciclo while

while indice <= limite:
  llave = claves[indice]
  suma += vocales[llave]
  lista.append(llave)
  indice += 1

suma
````

**2. Imprime todos los contactos de la agenda:**

````python
# Mostrar todos los registros de la agenda
contador = 1

for contacto in agenda:
  print(f'Contacto {contador}')
  print(f'Nombre: {contacto["nombre"]}')
  print(f'Teléfono: {contacto["telefono"]}')
  print(f'Correo: {contacto["correo"]}')
  print()
  contador += 1
````

**3. Crea un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre, teléfono y correo para un contacto.
   * Guarda estos datos en un diccionario llamado contacto.
   * Imprime el mensaje "Contacto creado con éxito".

````python
# Crear contacto
nombre = input('Inserta el nombre: ')
telefono = input('Inserta el teléfono: ')
correo = input('Inserta el correo: ')

contador = {
  "nombre":  nombre,
  "telefono": telefono,
  "correo": correo
}

agenda.append(contador)
print('Contacto creado con éxito')
print(agenda)
````

**4. Consulta un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre del contacto a consultar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces imprime el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".

````python
# Consultar contacto
buscar = input('Inserta el nombre a buscar: ')
encontrar = False

for contacto in agenda:
  if contacto['nombre'] == buscar:
    print("Contacto encontrado: ")
    print(contacto)
    encontrador = True

if encontrar == False:
  print('Contacto no encontrado')
````

**5. Actualiza un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre del contacto a actualizar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces imprime el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".
   * Crea un menú a través de un ciclo while que muestre las siguientes opciones:
   
   ```
   Menú actualizar contacto:
   
   1. Actualizar nombre
   2. Actualizar teléfono
   3. Actualziar correo
   4. Salir del menú
   ```
   * Solicita al usuario una opción mostrada en el menú, si el usuario no coloca la opción correcta, entonces muestra el mensaje "Contacto no encontrado".
   * Dada la opción escojida para modificar, ingresa ya sea el nuevo nombre, el nuevo teléfono o el nuevo correo del contacto encontrado.
   * Por cada registro nuevo ingresado, imprime de nueva forma el contacto para visualizar los cambios.
   * Define en la última opción un `break` para salir del menú.

````python
# Actualizar contacto
buscar = input('Inserta el nombre a actualizar: ')
encontrar = False

for contacto in agenda:
  if contacto['nombre'] == buscar:
    print("Contacto encontrado: ")
    print(contacto)
    encontrar = True

    bandera= True

    while bandera:
      print("Menú actualizar contacto:")
      print("1. Actualizar nombre")
      print("2. Actualizar teléfono")
      print("3. Actualizar correo")
      print("4. Salir del menú")

      opcion = int(input("Selecciona una opción: "))

      if opcion == 1:
        nuevo_nombre = input("Ingresa el nuevo nombre: ")
        contacto['nombre'] = nuevo_nombre
        print("Nombre actualizado con éxito.")
        print(contacto)
      elif opcion == 2:
        nuevo_telefono = input("Ingresa el nuevo teléfono: ")
        contacto['telefono'] = nuevo_telefono
        print("Teléfono actualizado con éxito.")
        print(contacto)
      elif opcion == 3:
        nuevo_correo = input("Ingresa el nuevo correo: ")
        contacto['correo'] = nuevo_correo
        print("Correo actualizado con éxito.")
        print(contacto)
      elif opcion == 4:
        print("Saliendo del menú de actualización de contacto.")
        bandera = False
      else:
        print("Opción no válida. Por favor, selecciona una opción válida.")
        if encontrar == False:
          print('Contacto no encontrado')
````

**6. Elimina un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre del contacto a eliminar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces elimina el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".

````python
# Eliminar contacto
buscar = input('Inserta el nombre a eliminar: ')
encontrar = False

for contacto in agenda:
  if contacto['nombre'] == buscar:
    agenda.remove(contacto)
    print("Contacto eliminado con éxito.")
    encontrar = True

    if encontrar == False:
      print('Contacto no encontrado')

    print(agenda)
````
